```python
import inspect
import random
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import basinhopping
```

### Умова завдання

```python
A, B = 2, 9
F = lambda x,y: (x**2 + y**2)/A - B*(np.cos(x) + np.cos(y))
x_limits = (-8, 8)
y_limits = (-8, 8)
```

### Графік заданої функції

```python
x_values = np.linspace(*x_limits, 100)
y_values = np.linspace(*y_limits, 100)
plt.figure(figsize=(10, 10))
X, Y = np.meshgrid(x_values, y_values)
Z = F(X, Y)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('F(x, y)')
ax.set_title('F(x, y)')

plt.show()
```

## Простий стохастичний пошук

```python
def simple_stochastic_search(function, x_bounds, y_bounds, step=0.1, iterations=1000):
    x_low, x_high = x_bounds
    y_low, y_high = y_bounds
    x = random.uniform(x_low, x_high)
    y = random.uniform(y_low, y_high)
    best_x = x
    best_y = y
    best_result = function(x, y)
    for _ in range(iterations):
        x = random.uniform(x_low, x_high)
        y = random.uniform(y_low, y_high)
        result = function(x, y)
        if result < best_result:
            best_result = result
            best_x = x
            best_y = y

    return best_x, best_y, best_result, iterations
```

```python
x_min, y_min, f_min, iterations = simple_stochastic_search(F, x_limits, y_limits, iterations=10000)
print(f'f_min = f({x_min}, {y_min}) = {f_min} made in {iterations} iterations')
```

## Імітація відпалу

```python
objective_function = lambda variables: F(*variables)
```

```python
class CustomStep:
    def __init__(self, temp_func):
        self.temp_func = temp_func
        self.iteration = 0

    def __call__(self, x):
        temp = self.temp_func(self.iteration)
        self.iteration += 1
        return x + temp * np.random.randn(*x.shape)
```

```python
def temp_func(iteration):
    return 1000.0 / (1 + iteration * 0.1)

result = basinhopping(
    objective_function,
    x0=np.array([0, 0]),
    minimizer_kwargs={"method": "L-BFGS-B", "bounds": [x_limits, y_limits]},
    take_step=CustomStep(temp_func),
    niter=1000
)
x_min, y_min, f_min = result.x[0], result.x[1], result.fun
print(f'f_min = f({x_min}, {y_min}) = {f_min} made in {result.nit} iterations')
```

## Порівняння часу виконання

```python
sss_results = {}
sim_annealing_results = {}

def sss_adapter(iter):
    x_min, y_min, f_min, iterations = simple_stochastic_search(F, x_limits, y_limits, iterations=iter)
    return x_min, y_min, f_min, iterations

def dual_annealing_adapter(iter):
    result = basinhopping(
        objective_function,
        x0=np.array([7/2, -17/4]),
        disp=False,
        minimizer_kwargs={"method": "L-BFGS-B", "bounds": [x_limits, y_limits]},
        take_step=CustomStep(temp_func),
        niter=iter,
    )

    it = result.lowest_optimization_result.nit

    return result.x[0], result.x[1], result.fun, it

def search_time_adapter(func, iter):
    start = time.time()
    x_min, y_min, f_min, iterations = func(iter)
    end = time.time()
    execTime = end - start
    return {
        "x_min": x_min,
        "y_min": y_min,
        "f_min": f_min,
        # "iterations": iterations,
        "execTime": execTime
    }

for iter_exp in range(0, 5):
    iters = 10 ** iter_exp
    sss_result = search_time_adapter(sss_adapter, iters)
    sss_results[iters] = sss_result

    res = search_time_adapter(dual_annealing_adapter, iters)
    sim_annealing_results[iters] = res

sss_df = pd.DataFrame.from_dict(sss_results, orient='index')
sim_annealing_df = pd.DataFrame.from_dict(sim_annealing_results, orient='index')

sss_df = sss_df.add_prefix('sss_')
sim_annealing_df = sim_annealing_df.add_prefix('sa_')

combined_df = pd.merge(sss_df, sim_annealing_df, left_index=True, right_index=True)

combined_df
```

## Порівняння різних функцій зміни температури

```python
funcs = [
    lambda i: 1000.0 / (1 + i * 0.1),
    lambda i: 1000.0 / (1 + i),
    lambda i: 1000.0 / (1 + i ** 2),
    lambda i: 1000.0 / (1 + np.sqrt(i)),
]

def dual_annealing_adapter_with_temp_func(temp_func):
    def adapter(iter):
        result = basinhopping(
            objective_function,
            x0=np.array([1, 0]),
            minimizer_kwargs={"method": "L-BFGS-B", "bounds": [x_limits, y_limits]},
            take_step=CustomStep(temp_func),
            niter=iter
        )
        return result.x[0], result.x[1], result.fun, result.nit
    return adapter

results = {}

for temp_func in funcs:
    func_str = inspect.getsource(temp_func)
    results[func_str] = search_time_adapter(dual_annealing_adapter_with_temp_func(temp_func), 1000)

results_df = pd.DataFrame.from_dict(results, orient='index')
results_df
```
