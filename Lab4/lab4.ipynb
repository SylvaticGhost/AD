## FP-Growth Algorithm

```python
transaction = df.apply(lambda row: [str(item) for item in row if pd.notna(item)], axis=1).tolist()
te = TransactionEncoder()
te_array = te.fit_transform(transaction)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
df_encoded.head(10)
```

```python
# This line uses the correct variable
frequent_itemsets = fpgrowth(df_encoded, min_support=MIN_SUPPORT, use_colnames=True)
frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(lambda x: list(x))
frequent_itemsets
```

```python
frequent_itemsets=association_rules(frequent_itemsets, metric="lift", min_threshold=1)
frequent_itemsets['antecedents'] = frequent_itemsets['antecedents'].apply(lambda x: list(x))
frequent_itemsets['consequents'] = frequent_itemsets['consequents'].apply(lambda x: list(x))
frequent_itemsets.sort_values("confidence",ascending=False)
frequent_itemsets
```

```python
rule_labels = [f"{', '.join(ant)} → {', '.join(con)}" for ant, con in
               zip(frequent_itemsets['antecedents'], frequent_itemsets['consequents'])]

plt.figure(figsize=(12, 8))

plt.scatter(range(len(rule_labels)), frequent_itemsets['confidence'],
            s=frequent_itemsets['support']*5000, alpha=0.6,
            c=frequent_itemsets['conviction'], cmap='viridis')
plt.colorbar(label='Переконливість (Conviction)')

plt.xticks(range(len(rule_labels)), rule_labels, rotation=90)
plt.ylabel('Достовірність (Confidence)')
plt.xlabel('Правила')
plt.title('Асоціативні правила')
plt.tight_layout()
plt.show()
```

```python
frequent_itemsets['antecedents_str'] = frequent_itemsets['antecedents'].apply(lambda x: ', '.join(x))
frequent_itemsets['consequents_str'] = frequent_itemsets['consequents'].apply(lambda x: ', '.join(x))

matrix_df = frequent_itemsets.pivot(index='antecedents_str', columns='consequents_str', values='conviction')

matrix_df.index.name = 'Antecedents'
matrix_df.columns.name = 'Consequents'

plt.figure(figsize=(12, 8))
sns.heatmap(matrix_df, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Асоціативні правила')
plt.tight_layout()
plt.show()
```

## Apriori алгоритм

```python
rules = apriori(transactions = transaction, min_support = MIN_SUPPORT, min_confidence = 0.1, min_lift = 1.01, min_length = 2, max_length = 2)
apriori_res = list(rules)

def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

aprioriDf = pd.DataFrame(inspect(apriori_res), columns = ['antecedents', 'consequents', 'Support', 'Confidence', 'Lift'])

aprioriDf
```
