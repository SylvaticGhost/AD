```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth, apriori
from mlxtend.preprocessing import TransactionEncoder
from eclat import eclat_algorithm
from helpers import sets_to_lists
from rules_vizualization import graph_rep, heat_map
from stat_report import stat_report
```

```python
MIN_SUPPORT = 0.1/2
```

## Дата сет

```python
df = pd.read_csv('./Market_Basket_Optimisation.csv')
df.head(8)
```

```python
df.shape
```

### Визначаємо частоту кожного продукту

```python
transaction = df.stack().values
transaction
```

```python
items = pd.DataFrame(transaction, columns=['product'])
items = items.dropna()

table = items.groupby("product").size().reset_index(name='count')
table.head(10)
```

```python
not_used_stats = ['antecedent support', 'consequent support', 'leverage', 'representativity', 'leverage', 'zhangs_metric', 'jaccard', 'certainty', 'kulczynski']
```

### Формуємо булеву таблицю відповідності товару до транзакції

```python
transaction = df.apply(lambda row: [str(item) for item in row if pd.notna(item)], axis=1).tolist()
te = TransactionEncoder()
te_array = te.fit_transform(transaction)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
df_encoded.head(10)
```

## FP-Growth алгоритм

```python
fpgrow_rules = fpgrowth(df_encoded, min_support=MIN_SUPPORT, use_colnames=True)
sets_to_lists(fpgrow_rules['itemsets'])
```

```python
fpgrow_rules=association_rules(fpgrow_rules, metric="lift", min_threshold=1)

sets_to_lists(fpgrow_rules['antecedents'])
sets_to_lists(fpgrow_rules['consequents'])

fpgrow_rules.drop(not_used_stats, axis=1, inplace=True)
fpgrow_rules = fpgrow_rules.sort_values('confidence', ascending=False)

fpgrow_rules
```

### Візуалізація результатів FP-Growth алгоритму

```python
graph_rep(fpgrow_rules, 'FP-Growth')
heat_map(fpgrow_rules, 'FP-Growth')
```

## Apriori алгоритм

```python
apriori_results = apriori(df_encoded, min_support=MIN_SUPPORT, use_colnames=True)
apriori_rules = association_rules(apriori_results, metric="lift", min_threshold=1.01)

sets_to_lists(apriori_rules['antecedents'])
sets_to_lists(apriori_rules['consequents'])

apriori_rules.drop(not_used_stats, axis=1, inplace=True)
apriori_rules = apriori_rules.sort_values("confidence", ascending=False)

apriori_rules
```

### Візуалізація результатів Apriori алгоритму

```python
graph_rep(apriori_rules, 'Apriori')
heat_map(apriori_rules, 'Apriori')
```

## ECLAT алгоритм

```python
_, eclat_rules = eclat_algorithm(df_encoded, MIN_SUPPORT, min_confidence=0.11)

eclat_rules_df = pd.DataFrame([(list(ant), list(con), sup, conf, lift, conv)
                              for ant, con, sup, conf, lift, conv in eclat_rules],
                             columns=['antecedents', 'consequents', 'support', 'confidence', 'lift', 'conviction'])

eclat_rules_df = eclat_rules_df.sort_values('confidence', ascending=False)
eclat_rules_df
```

### Візуалізація результатів ECLAT алгоритму

```python
graph_rep(eclat_rules_df, 'ECL')
heat_map(eclat_rules_df, 'ECL')
```

## Порівняння результатів

```python
stat_report([
    ('FP-Growth', fpgrow_rules),
    ('Apriori', apriori_rules),
    ('ECL', eclat_rules_df)
])
```
